```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
table = pd.read_excel(r'..\Data\Miudinho Mercado.xlsx')
```

## Cleaning the data

```python
table = table[:465]
```

```python
table['Codigo'] = table['Codigo'].astype(str)
```

```python
table
```

```python
# confirming that we don't have nan's
null0 = len(table['Concorrencia'][table['Concorrencia'].isna() == True])
notnull0 = len(table['Concorrencia'][table['Concorrencia'].isna() == False])
total = len(table['Concorrencia'])
```

```python
print(null0, notnull0, total)
print('null: {:,.2%} \nnotnull: {:,.2%}'.format(null0/total, notnull0/total))
```

## Analysing the data

```python
# Seeing how many competitor we have advertising on Google
'''
We put one less in calculation because of the 'nonexistent' competitor
'''
num_compet = len(table['Concorrencia'].unique())
print('In total, {} companies sell the same products as we do.'.format(num_compet-1)) 
```

```python
# Getting only the columns that we really need
analysis = table[['Codigo', 'Descricao', 'P.Venda', 'EAN', 'Marca', 'Produto', 'Preco Concorrencia','Concorrencia']]
```

```python
# Creating a new column to see which company has the better price in the internet
for i in range(len(analysis)):
    if analysis['P.Venda'][i] > analysis['Preco Concorrencia'][i]:
        analysis['menor_preco'][i] = analysis['Concorrencia'][i]
    if analysis['P.Venda'][i] < analysis['Preco Concorrencia'][i]:
        analysis['menor_preco'][i] = 'WeWin'
```

```python
# % difference
for i in range(len(analysis)):
    analysis['% diferenca'][i] =  round(((analysis['Preco Concorrencia'][i] / analysis['P.Venda'][i]) - 1), 2)
```

```python
analysis
```

```python
# Seeing how many times we had the better price and which companies are real competitors
menor_preco = analysis['menor_preco'].value_counts()
menor_preco
```

```python
lst_idx = analysis['menor_preco'].value_counts().index.tolist()
```

```python
# Seeing how many competitor really compete with our prices
real_compet = len(lst_idx)
print('In total, {} companies sell the same products as we do.'.format(num_compet -1))
print('But only {} companies have a better price than us in at least one product.'.format(real_compet)) 
```

### What is the percentage of times that we have the better price?

```python
for i in range(len(lst_idx)):
    print('{}: {:,.2%}'.format(lst_idx[i], menor_preco[i]/len(analysis)))
    print('-'*45)
```

## Analysing by brand

It seems that our biggest competitor is Magalu. But is it really? Magalu has better prices concerning all brands? Magalu has better prices concerning all kinds of products? Let's see... 

```python
analysis.groupby(by='Marca')['menor_preco'].value_counts()
```

```python
menor_Cefisa = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Cefisa']
menor_Kuka = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Kuka']
menor_Lolly = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Lolly']
```

```python
idx_cef = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Cefisa'].index.tolist()
idx_kuka = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Kuka'].index.tolist()
idx_lolly = analysis.groupby(by='Marca')['menor_preco'].value_counts()['Lolly'].index.tolist()
```

```python
len_Cefisa = analysis['Marca'] == 'Cefisa'
len_Cefisa = len_Cefisa.sum()

len_Kuka = analysis['Marca'] == 'Kuka'
len_Kuka = len_Kuka.sum()

len_Lolly = analysis['Marca'] == 'Lolly'
len_Lolly = len_Lolly.sum()
```

```python
# Cefisa
print('------------------ Cefisa -------------------')
for i in range(len(idx_cef)):
    print('{}: {:,.2%}'.format(idx_cef[i], menor_Cefisa[i]/len_Cefisa))
    print('-'*45)
```

```python
# Kuka
print('------------------ Kuka -------------------')
for i in range(len(idx_cef)):
    print('{}: {:,.2%}'.format(idx_kuka[i], menor_Kuka[i]/len_Kuka))
    print('-'*45)
```

```python
# Lolly
print('------------------ Lolly -------------------')
for i in range(len(idx_cef)):
    print('{}: {:,.2%}'.format(idx_lolly[i], menor_Lolly[i]/len_Lolly))
    print('-'*45)
```

Through this analysis we can see that Magalu is a problem only with Lolly products and it looks like that we have to look to our margin in Lolly products or that the competition is negotiating better prices.

The analysis concerning Kuka is that maybe some specific products are in promotion for some reason, but it seems that we are well placed concerning our prices.

In Cefisas case we need to look a little further in the products in which Mercado Livre has the better price to undestand if this is a specific thing or if they got a better deal with our supplier for some minor products.
